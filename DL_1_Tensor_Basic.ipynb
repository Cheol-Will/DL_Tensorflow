cast는 shape이 안 맞을 때 자동으로 처리해준다<br>
expansion 해서 계산해줌

```python
t1 = tf.random.normal(shape = (3, 4), mean = 0, stddev = 5)
t2 = tf.random.normal(shape = (1, 4), mean = 0, stddev = 5)

t1 = tf.cast(t1, dtype = tf.int16)
t2 = tf.cast(t2, dtype = tf.int16)

print(t1.numpy())
print(t2.numpy())
print('\n')
print(t1+t2)
```

```python
t1 = tf.random.normal(shape = (3, 4), mean = 0, stddev = 5)
t1 = tf.cast(t1, dtype = tf.int16)

t2 = tf.reduce_sum(t1)
print(t1.numpy())
print(t2.numpy())
```

```python
t2 = tf.reduce_sum(t1, axis = 0)
print(t1.numpy())
print(t2.numpy())
```

아래 내용들은 cnn의 이미지(rgb)에서 채널와이즈로 계산을 할 수 있게 해준다

```python
t1 = tf.random.normal(shape = (128, 128, 3), mean = 0, stddev = 5)
t1 = tf.cast(t1, dtype = tf.int16)

t2 = tf.reduce_sum(t1, axis = 2)
print(t1.numpy(), '\n')
print(t2.numpy()
```
